```python
open_ports = [22, 80, 443, 3389]

print("Starting Port Scan...\n")

for port in open_ports:
    print(f"Port {port} is OPEN")
```

```python
services = {
    22: "SSH",
    80: "HTTP",
    443: "HTTPS",
    3389: "RDP"
}

print("Detected Services\n")

for port in services:
    print(f"Port {port}: {services[port]}")
```

```python
high_risk_ports = [22, 3389]

print("Security Findings\n")

for port in high_risk_ports:
    print(f"Warning: Port {port} may require additional security monitoring")
```
